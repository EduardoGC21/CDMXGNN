```python
import pandas as pd

# Read only the last sheet from the Excel file
excel_path = "Raw/scian_2023_categorias_y_productos.xlsx"
last_sheet_name = pd.ExcelFile(excel_path).sheet_names[-1]
df_last_sheet = pd.read_excel(excel_path, sheet_name=last_sheet_name)
```

```python
df_last_sheet
```

```python
df_last_sheet = df_last_sheet[df_last_sheet.iloc[:, 0].notna()]
```

```python
df_last_sheet
```

```python
df_last_sheet = df_last_sheet.loc[1:].reset_index(drop=True)
```

```python
df_last_sheet.rename(columns={'Sistema de Clasificación Industrial de América del Norte, México SCIAN 2023': 'Codigo', 'Unnamed: 1': 'Descripcion'}, inplace=True)
df_last_sheet = df_last_sheet[['Codigo', 'Descripcion']]
df_last_sheet
```

```python
from unidecode import unidecode
import json

df_last_sheet['Descripcion'] = df_last_sheet['Descripcion'].apply(lambda x: unidecode(str(x)).lower())
codigo_descripcion_dict = dict(zip(df_last_sheet['Codigo'], df_last_sheet['Descripcion']))

json_str = json.dumps(codigo_descripcion_dict, ensure_ascii=False, indent=2)
print(json_str)
```

```python
with open(f"{last_sheet_name}.json", "w", encoding="utf-8") as f:
    f.write(json_str)
```
