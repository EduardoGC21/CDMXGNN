## Clean

```python
import geopandas as gpd
import pandas as pd
import re
from unidecode import unidecode

def process_crime_data(df):
    """
    Cleans, processes, and converts crime data from a pandas DataFrame to a GeoDataFrame,
    keeping only records from 2025 onwards or those with no date.
    """
    # Standardize column names to handle variations
    df.columns = [
        re.sub(r'_+', '_', unidecode(col).lower().replace(' ', '_').replace('-', '_').strip())
        for col in df.columns
    ]

    # Rename columns to a standard format
    columns_mapping = {
        'id_ci': 'id_ci',
        'fecha_de_los_hechos': 'fecha_de_los_hechos',
        'hora_de_los_hechos': 'hora_de_los_hechos',
        'alcaldia_hechos': 'alcaldia_hechos',
        'colonia_hechos': 'colonia_hechos',
        'calle_1_hechos': 'calle_1_hechos',
        'calle_2_hechos': 'calle_2_hechos',
        'tipo_impacto': 'tipo_impacto',
        'delito': 'delito',
        'modalidad_delito': 'modalidad_delito',
        'coord._x': 'coord_x',
        'coord_x': 'coord_x',
        'coord._y': 'coord_y',
        'coord_y': 'coord_y'
    }
    df = df.rename(columns=columns_mapping)
    
    # Ensure all required columns are present
    required_columns = [
        'id_ci','fecha_de_los_hechos', 'hora_de_los_hechos', 'alcaldia_hechos', 'colonia_hechos',
        'calle_1_hechos', 'calle_2_hechos', 'tipo_impacto', 'delito', 'modalidad_delito',
        'coord_x', 'coord_y'
    ]
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"Warning: Missing columns {missing_columns}. Skipping this DataFrame.")
        return None

    df = df[required_columns]

    print(f"Initial DataFrame shape: {df.shape}")

    # Normalize string columns
    for col in df.select_dtypes(include='object').columns:
        if col not in ['coord_x', 'coord_y']:
            df[col] = df[col].apply(lambda x: re.sub(r'_+', '_', unidecode(x.lower())) if isinstance(x, str) else x)

    # Convert date and time columns
    df['fecha_de_los_hechos'] = pd.to_datetime(df['fecha_de_los_hechos'], errors='coerce')
    
    # Filter to only keep records from 2025 onwards or with a null date
    df = df[(df['fecha_de_los_hechos'] >= '2025-01-01') | (df['fecha_de_los_hechos'].isna())]
    
    df['hora_de_los_hechos'] = pd.to_datetime(df['hora_de_los_hechos'], format='%H:%M', errors='coerce').dt.time
    
    # Clean and convert coordinate columns to a numeric type
    df['coord_x'] = pd.to_numeric(df['coord_x'].replace('sin registro', pd.NA), errors='coerce')
    df['coord_y'] = pd.to_numeric(df['coord_y'].replace('sin registro', pd.NA), errors='coerce')

    # Drop rows with any missing coordinate values
    df.dropna(subset=['coord_x', 'coord_y'], inplace=True)
    print(f"Shape after dropping NaN coordinates: {df.shape}")

    if df.empty:
        print("Warning: DataFrame is empty after filtering. Returning an empty GeoDataFrame.")
        return gpd.GeoDataFrame()

    # Create the GeoDataFrame
    gdf = gpd.GeoDataFrame(
        df,
        geometry=gpd.points_from_xy(df.coord_x, df.coord_y),
        crs="EPSG:6372"  # Set the Coordinate Reference System to SRID 6372
    )
    # Drop original coordinate columns
    gdf = gdf.drop(columns=['coord_x', 'coord_y'])
    
    # Order the final GeoDataFrame by date
    gdf = gdf.sort_values(by='fecha_de_los_hechos').reset_index(drop=True)
    
    return gdf
```

```python
gdf_list = []
for df in sheet_dfs.values():
    # Get the variable name of the DataFrame (e.g., AI_df, PG_df, FE_df)
    df_name = [name for name in globals() if globals()[name] is df][0]
    gdf_name = df_name.replace('_df', '_geo')
    gdf = process_crime_data(df)
    globals()[gdf_name] = gdf
    gdf_list.append(gdf)

```

```python
all_geo = gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True), crs="EPSG:6372")
all_geo = all_geo.to_crs("EPSG:6372")
all_geo
```

```python
all_geo = all_geo.rename(columns={'geometry': 'geom'})
```

```python
# Check for duplicates by all columns and drop them
if all_geo.duplicated().any():
    print(f"Found {all_geo.duplicated().sum()} duplicate rows. Dropping them.")
    all_geo = all_geo.drop_duplicates()

# Check for duplicates by 'id_ci' column and drop them
if 'id_ci' in all_geo.columns and all_geo.duplicated(subset=['id_ci']).any():
    print(f"Found {all_geo.duplicated(subset=['id_ci']).sum()} duplicate rows based on 'id_ci'. Dropping them.")
    all_geo = all_geo.drop_duplicates(subset=['id_ci'])

# Display the resulting DataFrame
# Reset the index of the DataFrame
all_geo = all_geo.reset_index(drop=True)
all_geo
```

## Split table

```python
all_geo['delito'].value_counts()
```

```python
# 1. Create the 'id' column from the DataFrame's index
all_geo['id'] = all_geo.index

# 2. Reorder the columns to place 'id' at the beginning (on the left)
cols = ['id'] + [col for col in all_geo.columns if col != 'id']
all_geo = all_geo[cols]

# 3. Create the points GeoDataFrame. 'id' is already the first column.
geo_points = all_geo[['id',  'delito', 'geom']]

# 4. Create the attributes DataFrame. 'id' will also be the first column here.
geo_attributes = all_geo.drop(columns=['geom', 'delito'])
```

## Save data

```python
import os

# Define the output directory
output_dir = '../../../0_DATA/1_CLEAN/CRIMEN/'

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# --- 1. Save the GeoDataFrame as a Shapefile (.shp) ---
# Note: Shapefiles truncate column names to 10 characters.
# 'tipo_impacto' will be saved as 'tipo_impa'.
points_filepath = os.path.join(output_dir, 'crime_points.parquet')
geo_points.to_parquet(points_filepath, index=False, engine='pyarrow')

# --- 2. Save the pandas DataFrame as a CSV file ---
# We use index=False because you already have a unique 'id' column.
attributes_filepath = os.path.join(output_dir, 'crime_attributes.parquet')
geo_attributes.to_parquet(attributes_filepath, index=False)
```
